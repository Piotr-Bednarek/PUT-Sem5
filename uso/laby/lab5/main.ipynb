```python
import numpy as np
from gekko import GEKKO
import matplotlib.pyplot as plt

# Parametry
n = 101

# Model
model = GEKKO()
model.options.IMODE = 6

# Parametry
model.time = np.linspace(0, 1, n)
t = model.Param(value=model.time)

# Ograniczenia
x = model.Var()
model.fix_initial(x, 1)
model.fix_final(x, 3)

# Inicjalizacja całki
J = model.Var(value=0)
model.fix_initial(J, 0)

# Równanie różniczkowe
model.Equation(J.dt() == 24 * x * t + 2 * x.dt()**2 - 4 * t)

# Cel
final_cost = model.FV()
final_cost.STATUS = 1
model.Connection(final_cost, J, pos2='end')
model.Obj(final_cost)

# Konfiguracja solvera
model.options.SOLVER = 3

# Rozwiązanie
model.solve(disp=False)

print(f"\nMinimalna wartość całki (GEKKO): {final_cost.value[0]:.2f}")
```

```python
import numpy as np
from gekko import GEKKO
import matplotlib.pyplot as plt

# --- 1. ROZWIĄZANIE PROBLEMU (GEKKO) ---
n = 101
model = GEKKO(remote=False)
model.options.IMODE = 6
model.options.NODES = 3

# Czas
model.time = np.linspace(0, 1, n)
t = model.Param(value=model.time)

# Zmienna x (z inicjalizacją liniową dla stabilności)
x_init = np.linspace(1, 3, n)
x = model.Var(value=x_init)

# Ograniczenia
model.fix_initial(x, 1)
model.fix_final(x, 3)

# Funkcjonał J
J = model.Var(value=0)
model.fix_initial(J, 0)

# Równanie różniczkowe
model.Equation(J.dt() == 24 * x * t + 2 * x.dt()**2 - 4 * t)

# Cel optymalizacji
final_cost = model.FV()
final_cost.STATUS = 1
model.Connection(final_cost, J, pos2='end')
model.Obj(final_cost)

# Solver
model.options.SOLVER = 3
model.solve(disp=False)

print(f"Minimalna wartość całki (GEKKO): {final_cost.value[0]:.4f}")

# --- 2. PRZYGOTOWANIE DANYCH ANALITYCZNYCH ---
t_arr = np.array(model.time)
x_analytical = t_arr**3 + t_arr + 1

# --- 3. RYSOWANIE WYKRESU (TWÓJ STYL) ---

# Ustawienie rozmiaru (zgodnie z Twoim kodem)
plt.figure(figsize=(12, 7))

# Rysowanie przebiegów (linewidth=2.5)
# Rozwiązanie analityczne - Niebieska ciągła
plt.plot(t_arr, x_analytical, 
         color='blue', linestyle='-', linewidth=2.5, 
         label='Analityczne: $x(t) = t^3 + t + 1$')

# Rozwiązanie numeryczne - Czerwona przerywana
plt.plot(model.time, x.value, 
         color='red', linestyle='--', linewidth=2.5, 
         label='Numeryczne (GEKKO)')

# Formatowanie tekstu (Title: 20pt Bold, Labels: 16pt)
plt.title('Porównanie rozwiązania numerycznego i analitycznego', 
          fontsize=20, fontweight='bold', pad=20)
plt.xlabel('Czas [s]', fontsize=16)
plt.ylabel('Wartość x(t)', fontsize=16)

# Dodatkowe elementy estetyczne
plt.legend(fontsize=14)  # Legenda też nieco większa, żeby pasowała
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
